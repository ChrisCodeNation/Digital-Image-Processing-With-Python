# 範例一

+ Detect one person in an image

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./lady.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, 'gray')
```

```python
# 載入已訓練好的人臉偵測權重
haar_cascade = cv2.CascadeClassifier('./haar_face.xml')

#偵測人臉
#return list：人臉的座標
face_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)

print(f'number of faces found = {len(face_rect)}')
print(f'coodinate of face = {face_rect[0]}')
```

```python
# 在影像上畫出bonfing box

for (x,y,w,h) in face_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
plt.imshow(img[:,:,::-1])
```

# 範例三
+ Detect a group of people in an image

Problem encountered:
+ 偵測不完全
+ 可能原因為：臉的部分區域被遮擋
    + 後排的人的臉沒有完全露出
    + 配戴眼鏡或帽子

```python
img = cv2.imread('./group 1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, 'gray')
```

```python
# 載入已訓練好的人臉偵測權重
haar_cascade = cv2.CascadeClassifier('./haar_face.xml')

#偵測人臉
#return list：人臉的座標
face_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)

print(f'number of faces found = {len(face_rect)}')

# 在影像上畫出bonfing box

for (x,y,w,h) in face_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
plt.imshow(img[:,:,::-1])
```
