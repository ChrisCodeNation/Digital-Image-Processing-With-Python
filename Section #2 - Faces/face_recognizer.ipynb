+ 使用face_train.py訓練完的權重來辨識

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

# load face detector
haar_cascade = cv2.CascadeClassifier('./haar_face.xml')

# load face reconizer
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
face_recognizer.read('./face_trained.yml')
```

```python
people = ['Ben Afflek', 'Elton John', 'Jerry Seinfield', 'Mindy Kaling', 'Madonna']

# load img
img = cv2.imread('./data/val/elton_john/1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# detect the face
face_rect = haar_cascade.detectMultiScale(gray, 1.1, 4)

for (x,y,w,h) in face_rect:
    face_roi = gray[y:y+h, x:x+w]
    label, confidence = face_recognizer.predict(face_roi)
    print(f'label = {people[label]} with the confidence of {confidence}')
    
    cv2.putText(img, str(people[label]), (20,20), cv2.FONT_HERSHEY_COMPLEX, 1.0, (0,255,0), thickness=2)
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), thickness=2)

plt.imshow(img[:,:,::-1])
```
