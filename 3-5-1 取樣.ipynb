```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Down_sampling(img, sampling_rate):
    nr, nc = img.shape
    # 取樣後的大小
    nr_s, nc_s = nr//sampling_rate, nc//sampling_rate 
    # 取樣後的影像
    result = np.zeros((nr_s, nc_s))
    
    for x in range(nr_s):
        for y in range(nc_s):
            result[x,y] = img[x*sampling_rate, y*sampling_rate]
                
    return result
```

```python
if __name__ == "__main__":
    img = cv2.imread("./lenna.png")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    nr, nc = img.shape[:2]
    # 取樣率
    sampling_rate = [1, 8, 16, 32, 64, 128] 
    plt.rcParams['figure.figsize'] = (8,8)
    
    for i, rate in enumerate(sampling_rate):
        plt.subplot(2, 3, i+1)
        plt.title(f'rate={str(rate)}')
        new_img = Down_sampling(img, rate)
        plt.axis('off')
        plt.imshow(new_img, cmap='gray')
    #plt.show()
```
