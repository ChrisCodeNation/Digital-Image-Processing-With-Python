# 一、Color spaces

+ GRAYSCALE
+ HSV
+ LAB

```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10,10)

plt.subplot(2,2,1)
img = cv2.imread('./cat.png')
plt.title('origin')
plt.imshow(img[:,:,::-1])

# BGR to grayscale
plt.subplot(2,2,2)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.title('gray')
plt.imshow(gray, cmap='gray')

# BGR to HSV
plt.subplot(2,2,3)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.title('hsv')
plt.imshow(hsv[:,:,::-1])

# BGR to LAB
plt.subplot(2,2,4)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
plt.title('lab')
plt.imshow(lab[:,:,::-1])
```

# 二、Blur technique

+ What is blur?
    +  blur is applied to the middle pixel as a result of the pixels around it
    + 利用一個小型的矩陣作為面罩（Mask）對原始影像進行捲積（Conolution），再將捲積得到值取代矩陣中心的像素值。
    + 通常用來消除或降低雜訊，例如camera sensor、燈光
     

+ [Average blur][1]
    + 計算 kernel裡所有 pixel的平均值，並將該平均值取代 kernel中心元素
    + kernel中的每個權pixel權重為1
    
+ [Gaussian blur][2]
    + 計算方式類似Average blur
    + kernel中的每個權pixel權重都不同，中心點最高，越往邊角越低
+ [Median blur][3]
    + 計算 kernel中所有 pixel的中位數(將pixel由小排到大)，並取然後取代中間的數值
+ [Bilateral blur][4]
    + 非線性濾波器
    + 由幾何空間距離和像素色差控制

[1]: https://medium.com/電腦視覺/影像雜訊去除-平滑法-smoothing-method-53ebf2b99570 "Average blur"
[2]: https://medium.com/@bob800530/python-gaussian-filter-概念與實作-676aac52ea17 "Gaussian blur"
[3]: https://medium.com/電腦視覺/影像雜訊去除-中值濾波器-median-filter-e00e1ec4c86d "Median blur"
[4]: https://cg2010studio.com/2012/10/14/雙邊濾波器-bilateral-filter/ "Bilateral blur"

```python
plt.rcParams['figure.figsize'] = (20,20)
plt.subplot(2,3,1)
img = cv2.imread('./cat.png')
plt.title('origin')
plt.imshow(img[:,:,::-1])

# averaging 
plt.subplot(2,3,2)
average = cv2.blur(img, (5,5))
plt.title('average blur')
plt.imshow(average[:,:,::-1])

# gaussian blur
plt.subplot(2,3,3)
gauss = cv2.GaussianBlur(img, (5,5), 0)
plt.title('gaussian blur')
plt.imshow(gauss[:,:,::-1])


#median blur
plt.subplot(2,3,4)
median = cv2.medianBlur(img, 5)
plt.title('median blur')
plt.imshow(median[:,:,::-1])

#bilateral
plt.subplot(2,3,5)
bilateral = cv2.bilateralFilter(img, 10, 30, 30)
plt.title('bilateral blur')
plt.imshow(bilateral[:,:,::-1])
```

# 三、BITWISE Operations

+ AND
+ OR
+ XOR
+ NOT

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

blank = np.zeros((400,400), dtype='uint8')

rectangle = cv2.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
circle = cv2.circle(blank.copy(), (200,200), 200, 255, -1)

plt.subplot(1,2,1)
plt.title('rectangle')
plt.imshow(rectangle, cmap='gray')

plt.subplot(1,2,2)
plt.title('circle')
plt.imshow(circle, cmap='gray')
```

```python
plt.rcParams['figure.figsize'] = (15, 15)

# bitwise AND
plt.subplot(1,4,1)
bitwise_and = cv2.bitwise_and(rectangle, circle)
plt.title('bitwise_and')
plt.imshow(bitwise_and, cmap='gray')

# bitwise_or
plt.subplot(1,4,2)
bitwise_or = cv2.bitwise_or(rectangle, circle)
plt.title('bitwise_or')
plt.imshow(bitwise_or, cmap='gray')

# bitwise_xor
plt.subplot(1,4,3)
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
plt.title('bitwise_xor')
plt.imshow(bitwise_xor, cmap='gray')

# bitwise_not(只輸入一張圖片)
plt.subplot(1,4,4)
bitwise_not = cv2.bitwise_not(rectangle)
plt.title('bitwise_not')
plt.imshow(bitwise_not, cmap='gray')
```
